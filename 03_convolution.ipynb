```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread("images/Lenna.png")
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
gray = rgb2gray(img)
```

```python
gray.shape
```

```python
plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
def conv2d(img, kernel):
    if kernel.shape != (3, 3):
        print("Kernel shape must be (3x3). Aborting.")
    
    if len(img.shape) != 2:
        print("Image must be grayscale. Aborting.")
    
    R_shape = (((img.shape[0] // 3) * 3) - 2, ((img.shape[1] // 3) * 3) - 2)
    R = np.zeros((R_shape[0], R_shape[1]))
    
    for i in range(R_shape[0]):
        for j in range(R_shape[1]):
            R[i, j] = np.sum(np.multiply(img[i:i+3, j:j+3], K))
    
    return R
```

```python
# Kernel to find vertical edges (Prewitt)
K = np.array([[-1, 0, 1],
              [-1, 0, 1],
              [-1, 0, 1]])
K
```

```python
R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Kernel to find horizontal edges (Prewitt)
K = np.array([[-1, -1, -1],
              [ 0,  0,  0],
              [ 1,  1,  1]])
K
```

```python
R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Kernel to find vertical edges (Sobel)
K = np.array([[-1, 0, 1],
              [-5, 0, 5],
              [-1, 0, 1]])
K
```

```python
R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Kernel to find horizontal edges (Sobel)
K = np.array([[-1, -5, -1],
              [ 0,  0,  0],
              [ 1,  5,  1]])
K
```

```python
R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Blur
K = np.array([[.0625, .125, .0625],
              [.125,   .25,  .125],
              [.0625, .125, .0625]])

R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Emboss
K = np.array([[-2, -1, 0],
              [-1,  1, 1],
              [ 0,  1, 2]])

R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Outline
K = np.array([[-1, -1, -1],
              [-1,  8, -1],
              [-1, -1, -1]])

R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```

```python
# Sharpen
K = np.array([[ 0, -1,  0],
              [-1,  5, -1],
              [ 0, -1,  0]])

R = conv2d(gray, K)

plt.imshow(R, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
plt.show()
```
