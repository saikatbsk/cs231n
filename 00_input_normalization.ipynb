```python
import numpy as np
import matplotlib.pyplot as plt

m = 100 # Number of samples

# Generating input
X1 = np.random.normal(17, 7, m)
X2 = np.random.normal(15, 5, m)
X = np.array([X1, X2]).T
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('input data')
plt.axhline(linewidth=4, color='r')
plt.axvline(linewidth=4, color='r')
plt.show()
```

```python
# Subtract mean
mean = np.sum(X, axis=0) / m
X = X - mean
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('input data')
plt.axhline(linewidth=4, color='r')
plt.axvline(linewidth=4, color='r')
plt.show()
```

```python
varience = np.sum(np.square(X), axis=0) / m
X /= varience
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('input data')
plt.axhline(linewidth=4, color='r')
plt.axvline(linewidth=4, color='r')
plt.show()
```
